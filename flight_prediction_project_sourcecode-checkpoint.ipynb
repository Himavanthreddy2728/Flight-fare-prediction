# Exploratory Data analysis

The feature "Date_of_Journey" needs to be converted into datetime format,then we extract corresponding day,month and year.

```python
df["Journey_day"] = pd.to_datetime(df["Date_of_Journey"],format="%d/%m/%Y").dt.day
df["Journey_month"] = pd.to_datetime(df["Date_of_Journey"],format="%d/%m/%Y").dt.month
df["Journey_year"] = pd.to_datetime(df["Date_of_Journey"],format="%d/%m/%Y").dt.year
```

Now we have to drop "Date_of_Journey" column

```python
df.drop(["Date_of_Journey"],axis=1,inplace=True)
df.head()
```

Similarly we have to perform the same on "Dep_Time","Arrival_Time" to extract corresponding hours and minutes respectively

```python
df["Dep_hours"]=pd.to_datetime(df["Dep_Time"]).dt.hour
df["Dep_minutes"]=pd.to_datetime(df["Dep_Time"]).dt.minute
df.drop(["Dep_Time"],axis=1,inplace=True)
df.head()
```

```python
df["Arrival_hours"]=pd.to_datetime(df["Arrival_Time"]).dt.hour
df["Arrival_minutes"]=pd.to_datetime(df["Arrival_Time"]).dt.minute
df.drop(["Arrival_Time"],axis=1,inplace=True)
df.head()
```

For "Duration":

```python
# Assigning and converting Duration column into list
duration = list(df["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:    # Check if duration contains only hour or mins
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"   # Adds 0 minute
        else:
            duration[i] = "0h " + duration[i]           # Adds 0 hour

duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))    # Extract hours from duration
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))
    
df["Duration_hours"] = duration_hours
df["Duration_mins"] = duration_mins

df.drop(["Duration"], axis = 1, inplace = True)
df.head()
```

```python
sns.catplot(y="Price",x="Destination",data=df.sort_values("Price",ascending=False),kind="boxen",height=6,aspect=3)
plt.show()
```

Hence "Destination" is also a nominal attribute,we perform OneHotEncoder using get_dummies()

```python
Destination=df["Destination"]
Destination=pd.get_dummies(Destination,drop_first=True)
Destination.head()
```

"Route" and "Total_stops" describe the same meaning, hence we can drop "Route" and use "Total_stops" as a feature.
Moreover "Price" is dependent on "Total_stops" and hence "Total_stops" is a ordinal attribute and we use "LabelEncoder"

```python
df.replace({"non-stop":0,"1 stop":1 ,"2 stops": 2,"3 stops":3,"4 stops": 4},inplace=True)
```

```python
df.drop(["Route"],axis=1,inplace=True)
df.head()
```

```python
df.drop(["Additional_Info"],axis=1,inplace=True)
df.head()
```

```python
x=df.loc[:,['Total_Stops', 'Journey_day', 'Journey_month', 'Journey_year',
       'Dep_hours', 'Dep_minutes', 'Arrival_hours', 'Arrival_minutes',
       'Duration_hours', 'Duration_mins', 'Airline_Air India', 'Airline_GoAir',
       'Airline_IndiGo', 'Airline_Jet Airways', 'Airline_Jet Airways Business',
       'Airline_Multiple carriers',
       'Airline_Multiple carriers Premium economy', 'Airline_SpiceJet',
       'Airline_Trujet', 'Airline_Vistara', 'Airline_Vistara Premium economy',
       'Source_Chennai', 'Source_Delhi', 'Source_Kolkata', 'Source_Mumbai',
       'Cochin', 'Delhi', 'Hyderabad', 'Kolkata', 'New Delhi']]
x.head()
```

# Extracting important features using ExtraTreesRegressor

```python
from sklearn.ensemble import ExtraTreesRegressor
selection= ExtraTreesRegressor()
selection.fit(x,y)
```

```python
print(selection.feature_importances_)
```

 # Visualization of Important features 

```python
plt.figure(figsize=(12,8))
feat_importances=pd.Series(selection.feature_importances_,index= x.columns)
feat_importances.nlargest(20).plot(kind='barh')
plt.show()
```

From the above plot, we can observe that "Total_Stops" is most important feature in predicting the "Price" and "Source_Mumbai" is least important feature involved to predict the "Price"

# Fitting a RandomForest model

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size= 0.8,random_state=3)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
#Predicting y values on x_test
y_pred=model.predict(x_test)
```

# Trainingset and Testingset scores

```python
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```

# Visualization of predicted variable y_pred with respect to actual variable y_test 

```python
plt.scatter(y_test,y_pred,alpha=1)
plt.xlabel="y_test"
plt.ylabel="y_pred"
plt.show()
```

```python
sns.distplot(y_test-y_pred)
plt.show()
```

The above distributionplot follows Gaussian Distribution and hence we can say our model is pretty good at predicting the values with a decent accuracy.

# Metrics

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
metrics.r2_score(y_test,y_pred)
```

Thus the coefficient of determination (regression score) achieved is 80.14%, now we perform hyperparameter tuning in order to increase the r2_score.

# Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
rf_model = RandomizedSearchCV(estimator = model, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_model.fit(x_train,y_train)
```

```python
rf_model.best_params_
```

```python
myprediction=rf_model.predict(x_test)
```

```python
plt.figure(figsize=(8,8))
sns.distplot(y_test-myprediction)
plt.show()
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, myprediction))
print('MSE:', metrics.mean_squared_error(y_test, myprediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, myprediction)))
```
